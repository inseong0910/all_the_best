```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []
```

```python
for row in data:
    if row[-1] != '' :
        result.append(float(row[-1]))
print(result)
```

```python
print(len(result))
```

```python
plt.plot(result, 'r')
plt.figure(figsize = (10, 2))
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = []
low = []
```

```python
for row in data :
    if row[-1] != '' and row[-2] != '' :
        if 1983 <= int(row[0].split('-')[0]) :
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14' :
                high.append(float(row[-1]))
                low.append(float(row[-2]))
            
plt.plot(high, 'hotpink')
plt.plot(low, 'skyblue')

plt.rc('font', family = 'Malgun Gothic')
plt.title('My Birthday weather')
plt.show()
```
